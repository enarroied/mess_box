```python
import time

def process_document_if_else(document_type):
    """
    Processes a document using if/else statements to determine the type.
    """
    if document_type == "pdf":
        return "Processing a PDF document: Extracting pages and text."
    elif document_type == "word":
        return "Processing a Word document: Converting to PDF for archival."
    elif document_type == "text":
        return "Processing a Plain Text document: Simple content read."
    else:
        return "Unknown document type: Cannot process."

# --- Benchmarking if/else ---
start_time_if_else = time.perf_counter()
for _ in range(100000): # Run many times for a measurable duration
    process_document_if_else("pdf")
    process_document_if_else("word")
    process_document_if_else("text")
    process_document_if_else("excel") # Test unknown
end_time_if_else = time.perf_counter()
duration_if_else = end_time_if_else - start_time_if_else

print("--- if/else Pattern ---")
print(process_document_if_else("pdf"))
print(process_document_if_else("word"))
print(process_document_if_else("text"))
print(process_document_if_else("excel"))
print(f"if/else Total Duration (100k iterations): {duration_if_else:.6f} seconds\n")
```

```python
# --- Polymorphism: Base Class and Concrete Implementations ---
class DocumentProcessor:
    """Abstract base class for all document processors."""
    def process(self):
        raise NotImplementedError("Subclasses must implement 'process' method.")

class PdfProcessor(DocumentProcessor):
    def process(self):
        return "Processing a PDF document: Extracting pages and text."

class WordProcessor(DocumentProcessor):
    def process(self):
        return "Processing a Word document: Converting to PDF for archival."

class TextProcessor(DocumentProcessor):
    def process(self):
        return "Processing a Plain Text document: Simple content read."

class UnknownProcessor(DocumentProcessor):
    def process(self):
        return "Unknown document type: Cannot process."

# --- Factory Pattern: Creates instances ---
class DocumentProcessorFactory:
    """
    Factory to create appropriate DocumentProcessor instances.
    """
    @staticmethod
    def get_processor(document_type):
        if document_type == "pdf":
            return PdfProcessor()
        elif document_type == "word":
            return WordProcessor()
        elif document_type == "text":
            return TextProcessor()
        else:
            return UnknownProcessor()

# --- Client Code: Uses the factory and polymorphism ---
def process_document_factory(document_type):
    processor = DocumentProcessorFactory.get_processor(document_type)
    return processor.process() # Polymorphic call

# --- Benchmarking Factory Pattern ---
start_time_factory = time.perf_counter()
for _ in range(100000): # Run many times for a measurable duration
    process_document_factory("pdf")
    process_document_factory("word")
    process_document_factory("text")
    process_document_factory("excel") # Test unknown
end_time_factory = time.perf_counter()
duration_factory = end_time_factory - start_time_factory

print("\n--- Factory Pattern with Polymorphism ---")
print(process_document_factory("pdf"))
print(process_document_factory("word"))
print(process_document_factory("text"))
print(process_document_factory("excel"))
print(f"Factory Total Duration (100k iterations): {duration_factory:.6f} seconds\n")
```
