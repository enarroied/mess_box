```python
def task1():
    while True:
        print("Task 1")
        yield


def task2():
    i = 0
    while True:
        print(f"Task {i}")
        i += 1
        yield


event_loop = [task1(), task2()]
```

```python
next(event_loop[1])
```

```python
import time


def sleep(seconds):
    start_time = time.time()
    while time.time() - start_time < seconds:
        yield


def task1():
    while True:
        print("Task 1")
        yield from sleep(1)


def task2():
    while True:
        print("Task 2")
        yield from sleep(5)


event_loop = [task1(), task2()]

while True:
    for task in event_loop:
        next(task)
```

```python
class Awaitable:
    def __await__(self):
        print("Inside __await__")
        yield "yielded"


async def main():
    awaitable_instance = Awaitable()
    result = await awaitable_instance
    print("Result from await:", result)
```

```python
coroutine = main()
```
