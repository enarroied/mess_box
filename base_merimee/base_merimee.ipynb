```python
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_merimee = pl.read_parquet("./data/merimee_full.parquet")
df_merimee.shape
```

```python
df_merimee.head(2)
```

```python
counts = df_merimee.group_by("region").len().sort("len", descending=True)

pdf = counts.to_pandas()

with plt.xkcd():
    plt.figure(figsize=(20, 16))
    sns.barplot(data=pdf, x="region", y="len", hue="region", palette="pastel")
    plt.xticks(rotation=90)
    plt.title("Nombre d'enregistrements par région")
    plt.xlabel("Région")
    plt.ylabel("Nombre de lignes")

    plt.legend(
        title="Région", loc="center left", bbox_to_anchor=(1.0, 0.5), borderaxespad=0
    )

    plt.tight_layout()
    plt.show()
```
